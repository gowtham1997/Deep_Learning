```python
from keras.datasets import imdb
```

```python
(X_train, y_train),(X_test, y_test) = imdb.load_data(num_words=10000) 
```

```python
import numpy
# one hot encoding reviews
def vectorise_reviews(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, seq in enumerate(sequences):
        results[i,seq] = 1
    return results
```

```python
X_train = vectorise_reviews(X_train)
X_test = vectorise_reviews(X_test)
```

```python
X_train[0]
```

```python
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
import keras.backend as K
def swish(x):
    return x * K.sigmoid(x)
```

```python
from keras.layers import Dense, Input
from keras.models import Model

inputs = Input(shape=(10000,))
x = Dense(16, activation=swish)(inputs)
x = Dense(16, activation=swish)(x)
prediction = Dense(1, activation='sigmoid')(x)
model = Model(inputs= inputs, outputs=prediction)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train,
                    y_train,
                    epochs=5,
                    batch_size=512,
                    validation_split=0.2)
```

```python
results = model.evaluate(X_test, y_test)
results
```

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
